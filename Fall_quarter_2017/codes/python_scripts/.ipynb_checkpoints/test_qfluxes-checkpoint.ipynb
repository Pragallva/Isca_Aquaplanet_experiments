```python
%matplotlib inline
import numpy as np
import netCDF4 as nc
import pylab as py

py.rc('text', usetex=True)
py.rc('font', family='serif', serif='Palatino',weight='bold')
```

```python
filename1="/project2/tas1/pragallva/Winter_quarter_2018/cmip3_data/qflux_O1.nc"
## project=CMIP3, model=CCSR/NIES/FRCGC, MIROC Model V3.2, high resolution, experiment=slab ocean control, ensemble=run1, time_frequency=mon, version=1 

filename2="/project2/tas1/pragallva/Winter_quarter_2018/cmip3_data/qflux_o1_slabcntl_1_cgcm3.1_t47.nc"
1. ## project=CMIP3, model=Canadian Centre for Climate Modelling and Analysis, CGCM3.1 Model, T63, experiment=slab ocean control, ensemble=run1, time_frequency=mon, version=1 

filename3="/project2/tas1/pragallva/Winter_quarter_2018/cmip3_data/qflux_O1.GISS1.Slabcntl.amended.nc"
2. ## project=CMIP3, model=NASA Goddard Institute for Space Studies, ModelE20/Russell, experiment=slab ocean control, ensemble=run1, time_frequency=mon, version=1 

filename4="/project2/tas1/pragallva/Winter_quarter_2018/cmip3_data/qflux_O1d_Slabcntl_run1.nc"
## project=CMIP3, model=CSIRO Atmospheric Research, Australia, Mk3.0 Model, experiment=slab ocean control, ensemble=run1, time_frequency=mon, version=1

filename5="/project2/tas1/pragallva/Winter_quarter_2018/cmip3_data/ocean_qflux.nc"
## Ruth's data
```

```python

data1=nc.Dataset(filename1,'r')
qflux1=data1.variables['qflux'][:]
ocean_flux1=qflux1.mean(axis=-1).mean(axis=0)
lat1=data1.variables['lat'][:]

data2=nc.Dataset(filename2,'r')
qflux2=data2.variables['qflux'][:]
ocean_flux2=qflux2.mean(axis=-1)      ### Does not have a seasonal cycle
lat2=data2.variables['lat'][:]

data3=nc.Dataset(filename3,'r')
qflux3=data3.variables['qflux'][:]
ocean_flux3=qflux3.mean(axis=-1).mean(axis=0)
lat3=data3.variables['lat'][:]

data4=nc.Dataset(filename4,'r')
qflux4=data4.variables['qflux'][:]
ocean_flux4=qflux4.mean(axis=-1).mean(axis=0)
lat4=data4.variables['lat'][:]

data5=nc.Dataset(filename5,'r')
qflux5=data5.variables['ocean_qflux'][:]
ocean_flux5=qflux5.mean(axis=-1).mean(axis=0)
lat5=data5.variables['lat'][:]
```

```python
a=6371e3
import scipy.integrate as integrate
def integrated(lat,x):
    l=np.deg2rad(lat)
    x=x*np.cos(l)
    int_x  =integrate.cumtrapz(x[::-1],l[::-1],axis=0,initial=None) #  (This is basically integration from - 90 deg)
    int_x_r=integrate.cumtrapz(x      ,l      ,axis=0,initial=None) #  (This is basically integration from + 90 deg) 
    avg_int_r=2*np.pi*a**2*(int_x[::-1][1:]+int_x_r[:-1])/2.0
    return avg_int_r/10**15

def zon_int(x):
    y=x*2*np.pi*np.cos(np.deg2rad(lat[1:-1]))*a
    return y/10**15

def plot_qflux_conv() :
    py.rc('text', usetex=True)
    py.rc('font', family='serif', serif='Palatino',weight='bold')
    py.plot(lat1,ocean_flux1,label='1',lw=2.5)
    py.plot(lat2,ocean_flux2,label='2',lw=2.5)
    py.plot(lat3,ocean_flux3,label='3',lw=2.5)
    py.plot(lat4,ocean_flux4,label='4',lw=2.5)
    py.legend(loc='best',fontsize=15,ncol=2)
    py.ylim(-60,60)
    py.xlim(-90,90)
    py.xticks(np.arange(-90,91,30))
    py.tick_params(labelsize=18,size=4,width=2)
    py.xlabel("latitude", fontsize=20)
    py.ylabel(r"q-flux conv in Wm^{-2}", fontsize=20)
# ocean_tranport=(-integrated(ocean_flux))
def plot_qflux() :
    py.plot(lat1[1:-1],(-integrated(lat1,ocean_flux1)),label='1',lw=2.5)
    py.plot(lat2[1:-1],(-integrated(lat2,ocean_flux2)),label='2',lw=2.5)
    py.plot(lat3[1:-1],(-integrated(lat3,ocean_flux3)),label='3',lw=2.5)
    py.plot(lat4[1:-1],(-integrated(lat4,ocean_flux4)),label='4',lw=2.5)
    py.legend(loc='best',fontsize=15,ncol=2)
    py.axhline(0,color='k')
    py.ylim(-4,4)
    py.xlim(-90,90)
    py.xticks(np.arange(-90,91,30))
    py.tick_params(labelsize=18,size=4,width=2)
    py.xlabel("latitude", fontsize=20)
    py.ylabel(r"q-flux in PW", fontsize=20)
```

```python
def plot_qflux_conv_Ruth() :
    py.rc('text', usetex=True)
    py.rc('font', family='serif', serif='Palatino',weight='bold')
    py.plot(lat5,ocean_flux5,label="Ruth's data",lw=2.5)
    py.legend(loc='best',fontsize=15,ncol=2)
    py.ylim(-60,60)
    py.xlim(-90,90)
    py.xticks(np.arange(-90,91,30))
    py.tick_params(labelsize=18,size=4,width=2)
    py.xlabel("latitude", fontsize=20)
    py.ylabel(r"q-flux conv in Wm^{-2}", fontsize=20)
def plot_qflux_Ruth() :
    py.plot(lat5[1:-1],(-integrated(lat5,ocean_flux5)),label="Ruth's data",lw=2.5)
    py.legend(loc='best',fontsize=15,ncol=2)
    py.axhline(0,color='k')
    py.ylim(-4,4)
    py.xlim(-90,90)
    py.xticks(np.arange(-90,91,30))
    py.tick_params(labelsize=18,size=4,width=2)
    py.xlabel("latitude", fontsize=20)
    py.ylabel(r"q-flux in PW", fontsize=20)
```
