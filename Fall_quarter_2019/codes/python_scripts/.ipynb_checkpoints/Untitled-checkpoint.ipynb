```python
dtheta=np.radians(lat[1]-lat[0])
R = 6371.e3
def spher_div_vert(x, ax=-1):
       N=50
       fac=1 #np.cos(np.radians(lat))
       div= smooth_multi( np.gradient((x)*(fac)/R,dtheta, axis=ax), N, smoothing=1,  ax=ax)
       return div

div_TE=spher_div_vert(TE)
div_SE=spher_div_vert(SE)
```

```python
def find_zero(div,n=1):
  if n>0 :
    l_range=((lat>=30) & (lat<=70)) 
  else :
    l_range=((lat>=-60) & (lat<=-30)) 
    
  zero_loc = np.zeros( (div.shape[0], div.shape[1], div.shape[2]) )
  for m in range(div.shape[0]):
      for mo in range(div.shape[1]):
         for ye in range(div.shape[2]):               
            interpolation_function = interp1d(div[m,mo,ye,l_range],lat[l_range],kind='linear')
            try :
                zero_loc[m,mo,ye]=interpolation_function(0)
            except ValueError as err:
                zero_loc[m,mo,ye]=np.nan        
  return zero_loc
```

```python
TE_zero_N=find_zero(div_TE,  1)
TE_zero_S=find_zero(div_TE, -1)

TESE_zero_N=find_zero(div_TE+div_SE,  1)
TESE_zero_S=find_zero(div_TE+div_SE, -1)
```

```python
def inerp_at_max(flux,lat,n=1):
  if n>0 :
    l_range=((lat>=30) & (lat<=70)) 
    storm_pos=TE_zero_N
  else :
    l_range=((lat>=-70) & (lat<=-30)) 
    storm_pos=TE_zero_S
  D=np.copy(storm_pos)
  for m in range(flux.shape[0]):
      for mo in range(flux.shape[1]):
         for ye in range(flux.shape[2]):               
             X = storm_pos
             interpolation_function = interp1d(lat[l_range],flux[m,mo,ye,l_range],kind='cubic')
             try :
                D[m,mo,ye]=interpolation_function(X[m,mo,ye])
             except ValueError as err:
                D[m,mo,ye]==np.nan        
  return D

Intensity_te_N=inerp_at_max(TE,lat,1)
Intensity_te_S=inerp_at_max(TE,lat,-1)

Intensity_tese_N=inerp_at_max(TE+SE,lat,1)
Intensity_tese_S=inerp_at_max(TE+SE,lat,-1)
```
